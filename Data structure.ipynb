```python
#Explain the key features of lists in python.

#Lists in Python are a versatile and widely used data structure.list is like a shopping bag which can store everything.list are mutable.
here are some key features:
1- Ordered collection:- list maintain the order of elements. each element can be accessed by its index, starting from 0.

2- Dynamic size:- lists can grow and shrink in size as needed you can add or remove elemens using methods..
append(), extend(), insert(), remove()

3- Comprehensive methods:- Lists come with a variety of built in methods for common operations.
sort(), reverse(), count(), pop() etc...

4- Nested lists:- Lists can contain other lists as elements enabling the creation of multi dimensional arrays or matrices.
```

```python
#Describe how to add, modify, and delete items in a dictionary with examples.

#lets dive into dictionary modification in python.
1- Adding items
my_dict ={'name': 'Suraj', 'age': 25}
my_dict['city'] = 'Uttar pradesh'
print(my_dict)
{'name': 'Suraj', 'age': 25, 'city': 'Uttar pradesh'}

2- Modifying items
my_dict['age'] = 26
print(my_dict)
{'name': 'Suraj', 'age': 26, 'city': 'Uttar pradesh'}

3- Deleting items
del my_dict['city']
print(my_dict)
{'name': 'Suraj', 'age': 26}
```
